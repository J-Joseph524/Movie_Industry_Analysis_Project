```python
import datetime 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
pd.set_option('display.max_columns', 1000)     #opening most recently saved dataframe
df5 = pd.read_csv('df5.csv', index_col = 0)
df5
```

```python
df5['first_air_date'][0]    #parsing down in 'first_air_date' column
```

```python
df5['first_air_date'] = pd.to_datetime(df5['first_air_date'], format='%Y/%m/%d')   #converting 'first_air_date' column to datetime type
df5
```

```python
filtered_df = df5[df5['first_air_date'] >= datetime.datetime(2000, 1, 1)].copy()   #filtering out tv shows that started before the year 2000
```

```python
filtered_df
```

```python
df6 = filtered_df.groupby('NBC')['vote_average'].mean()     #groupby one network to find average of 'vote_average'. 
df6
```

```python
df6[1]            #capturing average of 'vote_average' from above cell
```

```python
list_of_networks = ['ABC', 'NBC', 'Netflix', 'CBS', 'BBC One', 'FOX', 'ITV', 'BBC Two', 'TBS', 'HBO', 'Nickelodeon', 'Channel 4', 'MBS', 'Cartoon Network', 'SBS', 'Disney Channel', 'Amazon']
 #created top 17 list of network companies with most occurances
```

```python
network_vote_avg = []                     #creating a list of averages of 'vote_average' for each network
for i in list_of_networks:
    network_vote_avg.append(filtered_df.groupby(i)['vote_average'].mean()[1])

network_vote_avg
```

```python
fig = plt.figure(figsize=(18,8))               #creating bar graph to show vote average for each network
ax = fig.add_axes([0,0,1,1])
e = ['ABC', 'NBC', 'Netflix', 'CBS', 'BBC One', 'FOX', 'ITV', 'BBC Two', 'TBS', 'HBO', 'Nickelodeon', 'Channel 4', 'MBS', 'Cartoon Network', 'SBS', 'Disney Channel', 'Amazon']
plt.xlabel('Networks', fontsize = 28) 
plt.ylabel('Vote Average', fontsize = 28)
plt.title('Vote Average of Each Network', fontsize = 32)
plt.legend(e, loc=2, fontsize= 30)   #
ax.tick_params(axis='both', which='major', labelsize=20)
plt.xticks(rotation=70)
#plt.axes.tick_params(axis='both', labelsize=15)
ax.bar(e,network_vote_avg,color=['black', 'red', 'green', 'blue', 'cyan', "orange", "pink", "grey", "violet", "gold", "lime", "aqua", "navy", "teal", "yellow", "brown", "purple"])
plt.show()
```

```python
import ast                          #creating a new column to have runtime as a 
filtered_df['episode_run_time']=filtered_df['episode_run_time'].apply(lambda x: ast.literal_eval(x))
filtered_df.head(50)
```

```python
filtered_df['episode_run_time'][0][0]
```

```python
type(filtered_df['episode_run_time'][0])
```

```python
filtered_df['runtime'] = filtered_df.episode_run_time.apply(lambda x: x[0] if len(x) > 0 else np.nan)  
#creating a new column for runtime, using the first element of the list
```

```python
filtered_df.runtime.value_counts()   #checking to see number of occurances of each runtime
```

```python
filtered_df['n_runtime'] = pd.cut(filtered_df.runtime, bins=[0,30,60,90,120], labels= ['0-29', '30-59', '60-89', '89-119'])
#creating a new column to put runtimes into buckets of time ranges
```

```python
filtered_df
```

```python
y = filtered_df.groupby('n_runtime').mean()   #creating a new data frame to see shows grouped by the new buckets
```

```python
y            #using this data frame for the runtime bar graph
```

```python
fig = plt.figure(figsize=(18,8))               #creating bar graph to show vote average for each runtime range
ax = fig.add_axes([0,0,1,1])
plt.xlabel('Runtime Range', fontsize = 28) 
plt.ylabel('Vote Average', fontsize = 28)
plt.title('Vote Average of Runtime Range', fontsize = 32)
plt.legend(e, loc=2, fontsize= 30)   #
ax.tick_params(axis='both', which='major', labelsize=20)
ax.bar(y['n_runtime'],y['vote_average'],color=['lime', 'cyan', 'red', "pink"])
plt.show()
```
